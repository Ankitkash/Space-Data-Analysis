# Removal of Outliers

```python
def remove_outliers(col_name):
  sorted(col_name)
  Q1, Q3 = col_name.quantile([0.25, 0.75])
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  return lower, upper
```
